# 【测试】判断文件类型

pathlib是Python新推出的一个处理路径的模块，其使用了面向对象的思想来开发，所以使用的时候需要了解一定的类和对象的知识。

pathlib的相关说明可以[在此](https://docs.python.org/zh-cn/3/library/pathlib.html)查看

用pathlib很容易获取文件的扩展名：

````python
import pathlib
p = pathlib.Path('D:\test.txt')
print(p.suffix)
````

扩展名中包含了这个文件可能是什么类型文件的信息，例如jpg常常是图片文件，docx常常是word的文档。

接下去请用pathlib，基于给予的路径的后缀名，判断文件类型，主要分为以下几类：

- 图片
- 视频
- 音频
- 文档
- 软件
- 其它

不是常见的后缀名就归类为其它。后缀名的常见分类请自行上网检索。

例如：

#### 输入：

```
C:\Windows\System32\cmd.exe
```

#### 输出：

```
软件
```
