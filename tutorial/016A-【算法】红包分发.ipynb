# 【算法】红包分发

有一群小伙伴互相之间群发红包，略不同于微信的红包，发红包的人，给指定的几个人平分一个红包内的金额，如果无法分干净，那么多的就被发红包的人拿回。

例如A发红包给C、E、F，红包金额7元，那么C、E、F各自拿到2元，多出来的1元返回给A。

假设刚开始时每个人的资金账户都是0元，允许无限无息借贷（也就是说允许账户出现负数），请编写程序计算几轮过后每位小伙伴的账户余额。

每个输入的样本格式如下：

第一行为参与小伙伴的总人数n，

第二行到第n + 1行为这些小伙伴的名字。

第n + 2行为开始发红包的小伙伴名字，

第n + 3行为总红包的数量a1和发给几个人k1，用空格分割

第n + 4行到n + k1 + 4行开始为这些拿红包的人的名字

第n + k1 + 5行为第二次发红包相关的数据，以此类推

最后一行以**end**作为结尾

输出为n行，每一行为小伙伴的名字加账户余额，以空格隔开，名字保持输入时的顺序。

## 样本举例：

#### 输入：

```
5
dave
laura
owen
vick
amr
dave
200 3
laura
owen
vick
owen
500 1
dave
amr
150 2
vick
owen
laura
0 2
amr
vick
vick
0 0
end
```

#### 输出：

```
dave 302
laura 66
owen -359
vick 141
amr -150
```
